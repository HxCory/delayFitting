```python
import numpy as np
from scipy.interpolate import UnivariateSpline
import matplotlib.pyplot as plt
import pylab as plb
%matplotlib inline
```

```python
def gaussianPulse(t, E0, cen, T, omega, phi):
    return (E0) * np.exp(-8 * np.log(2) * (t-cen)**2 / (2*T**2)) * np.cos((omega * t) + phi) 
def FWHM(alpha, T):
    return T/np.sqrt(alpha)
```
