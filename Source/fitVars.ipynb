```python
from numpy import sqrt, pi, exp, linspace, loadtxt
import numpy as np
from lmfit import  Model
import os
import matplotlib.pyplot as plt
import pylab as plb
import definitionsFits as defs
import twoPhotonDipole as func
%matplotlib inline

Path = '/Users/cgoldsmith/Desktop/text_files_data'
os.chdir(Path)
data = loadtxt('TDSE_3fs.txt')
x = data[:, 0]
y = data[:, 4]
x2 = np.zeros(np.size(x) - 3)
y2 = np.zeros(np.size(x) - 3)
plt.plot(x, y,'bo')
# plt.plot(defs.omega_dip, func.fitDipoleFactors(defs.omega_dip, 1, 1, 1, 1))
j = 0
for i in x:
    if(i > 0.675):
        x2[j] = i
        j = j + 1
j = 0
m = 0
for i in y:
    if(m > 2):
        y2[j] = i
        j = j + 1
    m = m + 1
# print [x2, y2]
print x2
# # plt.show()
```

```python
gmod = Model(func.fitDipoleFactors)
result = gmod.fit(y2, omegaF = x2, m1 = -4.38012893, m3 = -16.96671107,\
	c_1 = 0.33987732,  c_3 = -0.21666822, free = 0.033361646)

print(result.fit_report())

plt.plot(x2, y2,'bo')
plt.plot(x2, result.init_fit, 'k--')
plt.plot(x2, result.best_fit, 'r-')
```
